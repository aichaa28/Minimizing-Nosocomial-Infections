```python
columns_to_normalize = ['PRE4', 'PRE5', 'AGE']

# Calculate means and variances of selected columns
means = df[columns_to_normalize].mean()
variances = df[columns_to_normalize].var()

# Normalize selected columns while preserving means and variances
df[columns_to_normalize] = (df[columns_to_normalize] - df[columns_to_normalize].min()) / (df[columns_to_normalize].max() - df[columns_to_normalize].min())

```

```python
X = df.drop(columns=['Risk1Yr'])
y = df['Risk1Yr']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print("Data type of y_train:", y_train.dtype)
print("Unique values of y_train:", y_train.unique())
```

```python
y_pred_prob = model.predict_proba(X_test)[:, 1]
```

```python
y_pred = (y_pred_prob >= 0.5).astype(int)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
```
